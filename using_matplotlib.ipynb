```python

import matplotlib.pyplot as plt
import numpy as np
import math
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show()
```

```python
import matplotlib.pyplot as plt
categories=['lion','horse','cat']
values=[23,45,67]
plt.bar(categories,values,color='pink')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('basic vertical bar graph')
plt.show()


```

```python
import matplotlib.pyplot as plt
categories=['ferrari','benz','honda']
values=[40,50,60]
plt.barh(categories,values,color=['green','blue','pink'])
plt.xlabel('distance covered/km')
plt.ylabel('car name')
plt.title('AR race 1')
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
categories=['lion','horse','cat']
values1=[23,45,67]
values2=[25,43,76]
bar_width=0.35
bar_position1=np.arange(len(categories))
bar_position1=np.arange(len(categories))
bar_position2=bar_position1+bar_width
plt.bar(bar_position1,values1,width=bar_width,label='2021',color='green')
plt.bar(bar_position2,values2,width=bar_width,label='2022',color='blue')
plt.xlabel('categories')
plt.ylabel('values')

```

```python

import matplotlib.pyplot as plt
categories=['transport','income tax','housing']
values1=[15,24,30]
values2=[20,18,25]
plt.bar(categories,values1,label='first half',color='skyblue')
plt.bar(categories,values2,bottom=values1,label='second half',color='pink')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('stacked bar graphs')
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
categories =['lion','horse','kangaroo']
values1=[15,25,56]
values2=[20,18,35]
bar_width=0.35
bar_positions1=np.arange(len(categories))
bar_positions2=bar_positions1+bar_width
plt.bar(bar_positions1,values1,width=bar_width,label='2021',color='skyblue')
plt.bar(bar_positions2,values2,width=bar_width,label='2022',color='orange')
plt.xlabel('categories')
plt.ylabel('values')
```
